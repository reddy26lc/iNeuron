### In this assignment students have to compress racoon grey scale image into 5 clusters. In the end, visualize both raw and compressed image and look for quality difference. The raw image is available in spicy.misc package with the name face.

```python
import numpy as np
from sklearn import cluster, datasets
from sklearn.cluster import KMeans
from scipy import misc
import matplotlib.pyplot as plt

image = misc.face(gray=True)
```

```python
# Compressing the gray scale image into 5 clusters
n_clusters = 5
np.random.seed(0)

X = image.reshape((-1,1))
kmeans = KMeans(n_clusters=n_clusters, n_init = 4)
kmeans.fit(X)
```

```python
values = kmeans.cluster_centers_.squeeze()
labels = kmeans.labels_

#Create an array from labels and values
X_compressed = np.choose(labels, values)
X_compressed.shape = image.shape
```

```python
vmin = image.min()
vmax = image.max()
```

```python
#Original image
plt.figure(1)
plt.imshow(image, cmap= plt.cm.gray, vmin=vmin, vmax = 256)
```

```python
#Compressed image
plt.figure(2)
plt.imshow(X_compressed, cmap = plt.cm.gray, vmin=vmin, vmax = vmax)
```

```python
plt.figure(3)
plt.imshow(X_compressed, cmap = plt.cm.gray, vmin=vmin, vmax = vmax)
plt.contour(X_compressed, [50, 200])
```
