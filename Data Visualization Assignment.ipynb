```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
iris_df = datasets.load_iris()
```

```python
X_reduced = PCA(n_components = 3).fit_transform(iris_df.data)
Y= iris_df.target

iris_fig = plt.figure(1, figsize=(8,8))
iris_plot = Axes3D(iris_fig)
colors = ['r','g','b']
iris_plot.scatter(X_reduced[:,0],X_reduced[:,1],X_reduced[:,2],c=Y,cmap=plt.cm.rainbow_r)

iris_plot.set_title("3D plot of Iris data")
iris_plot.set_xlabel("PCA component 1")
iris_plot.set_ylabel("PCA component 2")
iris_plot.set_zlabel("PCA component 3")

plt.show()
```
